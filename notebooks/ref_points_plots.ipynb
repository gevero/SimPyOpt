## Libraries

```python
# libraries
import sys
import numpy as np

# local utils
sys.path.append('../')
import utils

# for creating a responsive plot 
%matplotlib widget 
  
# importing required libraries 
from mpl_toolkits.mplot3d import Axes3D 
import matplotlib.pyplot as plt 
```

## compute reference points

```python
# ref points parameters
x0, y0,z0 = 10, 10, 5
n_u, n_v = 100, 100
u_scale, v_scale = 0.7,0.7

# compute reference points
ref_points = np.array(
    utils.nsga_ref_points(
        x0, y0, z0, n_u=n_u, n_v=n_v, u_scale=u_scale, v_scale=v_scale
    )
)
```

## Plot reference points

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection="3d")
ax.scatter(ref_points[:, 0], ref_points[:, 1], ref_points[:, 2])
ax.set_zlim(0, 10);
```
